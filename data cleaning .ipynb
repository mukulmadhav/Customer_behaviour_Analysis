```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename (columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
# create a column age group.
labels =['Young Adult','Adult','Middle_Aged','Senior']
df['age_group']= pd.qcut(df['age'], q=4 , labels = labels )
```

```python
df[['age','age_group']].head(10)
```

```python
df['frequency_of_purchases'].unique()
```

```python
#create a column purchase_frequency 
frequency_maping = {
     'Weekly':7,
     'Fortnightly':14,
     'Monthly':30,
     'Annually':365,
     'Bi-Weekly':14,
     'Quarterly':90,
     'Every 3 Months':90   
 }
df['purchase_frequency']=df['frequency_of_purchases'].map(frequency_maping )
```

```python
df[['frequency_of_purchases','purchase_frequency']].head(10)
```

```python
df=df.drop('promo_code_used',axis=1)
```
